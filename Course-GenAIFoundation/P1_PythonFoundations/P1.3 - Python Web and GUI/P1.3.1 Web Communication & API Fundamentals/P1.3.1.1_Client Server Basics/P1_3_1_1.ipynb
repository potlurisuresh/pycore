## ğŸŒ IP Address & Port Number
**IP address = address of the computer.**
It is a unique number that identifies a device on a network.

**Examples:**
- 192.168.1.10
- 142.250.182.14

**Port number = address of the application.**
One computer can run many apps. A port number identifies a specific application running on a machine.

**Examples :**  One Computer can run:
- Web server
- Database
- Email server
- Backend API

**How does the system know which program should handle the request?**
Using a port number.

**Simple real example:**
