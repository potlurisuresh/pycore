# Beginner Assessment 3: Solutions
Reference answers for all questions.

1. HTTP is a protocol for transferring web pages.

2. URL stands for Uniform Resource Locator.

3. 404 is the status code for Not Found.

4. HTTPS is the secure version of HTTP.

5. GET is used to retrieve data from a server.

6. 200 is the status code for success.

7. Query string is the part of a URL that contains data to be passed to web applications.

8. 80 is the default port for HTTP.

9. Web browser is a software application for accessing information on the web.

10. Content-Type header tells the server the type of data being sent.
