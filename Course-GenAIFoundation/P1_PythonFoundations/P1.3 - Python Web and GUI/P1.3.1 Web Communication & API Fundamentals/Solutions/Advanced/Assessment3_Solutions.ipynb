# Advanced Assessment 3: Solutions
Reference answers for all questions.

1. PATCH is used for partial updates to a resource.

2. Statelessness means each request contains all information needed; the server does not store client state.

3. 401 is Unauthorized; 403 is Forbidden.

4. CORS stands for Cross-Origin Resource Sharing; it controls access to resources from different domains.

5. OPTIONS describes communication options for the target resource.

6. An API endpoint is a specific URL where an API can be accessed. Example: /api/users

7. Accept header tells the server what content types the client can process.

8. PUT replaces the entire resource; PATCH updates part of it.

9. 204 means No Content; used when a request is successful but there is no content to return.

10. JSON is lightweight, easy to read, and widely supported.
