# Beginner Assessment 4: Matching
Match each HTTP method to its correct purpose.

1. GET      a) Retrieve data

2. POST     b) Create a new resource

3. PUT      c) Update a resource

4. DELETE   d) Delete a resource

5. PATCH    e) Partially update a resource

6. HEAD     f) Retrieve headers only

7. OPTIONS  g) Describe communication options

8. CONNECT  h) Establish a tunnel to the server

9. TRACE    i) Echo the received request

10. COPY    j) Duplicate a resource (if supported)
