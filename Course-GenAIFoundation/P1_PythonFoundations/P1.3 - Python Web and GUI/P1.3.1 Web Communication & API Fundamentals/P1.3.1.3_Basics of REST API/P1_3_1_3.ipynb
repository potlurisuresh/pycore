## ‚ùì What is an API?
API stands for **Application Programming Interface**.
It defines **how two software systems talk to each other** using requests and responses.

### ‚úÖ Why APIs Matter
- They let apps share data and functionality
- They hide internal complexity behind a clean interface
- They enable integration across services

## üîé Common API Types (REST is one of them)
There are multiple styles and technologies for APIs:
- **REST** (most common on the web)
- **SOAP** (older, XML-based, strict standards)
- **GraphQL** (client can request exactly what it needs)
- **gRPC** (fast, binary, service-to-service APIs)
- **WebSockets** (real-time two-way communication)

REST is popular because it is simple, readable, and built on standard HTTP.

## üß© What is REST?
**REST** stands for **Representational State Transfer**.
It is a style of designing web APIs so that clients and servers can communicate in a consistent, predictable way.

REST APIs are **stateless** ‚Äî each request contains all the information the server needs to process it.
The server does not remember previous requests from the client.

In REST, you interact with **resources** (data objects) using standard HTTP methods.

## üß± Key REST Concepts
### 1) Resources
A resource is a piece of data you can access or modify.
Examples: users, products, orders, posts

### 2) Endpoints
An endpoint is a URL that represents a resource.
Examples:
- /api/users
- /api/products/42

### 3) Representation
The resource is represented in a format like JSON or XML.
Most modern APIs use JSON.

## üîÅ REST and HTTP Methods
REST works *with* HTTP methods to describe actions on resources.

| Action | HTTP Method | Example |
|---|---|---|
| Read all users | GET | GET /api/users |
| Read one user | GET | GET /api/users/5 |
| Create user | POST | POST /api/users |
| Update user | PUT | PUT /api/users/5 |
| Delete user | DELETE | DELETE /api/users/5 |

## ‚úÖ Key Takeaways
- An API defines how systems communicate via requests and responses
- There are multiple API styles; REST is the most common on the web
- REST is a **design style** for building APIs
- REST APIs are **stateless** (each request stands alone)
- Resources are accessed via endpoints and represented as JSON
- HTTP methods describe actions on resources
