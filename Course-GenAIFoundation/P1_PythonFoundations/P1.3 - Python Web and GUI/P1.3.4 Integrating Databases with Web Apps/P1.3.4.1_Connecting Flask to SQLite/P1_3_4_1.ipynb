Flask is a popular Python web framework. It can use SQLite, a lightweight database, for storing data.

## üöÄ Flask: Connecting to SQLite
Flask uses the `sqlite3` module or extensions like `Flask-SQLAlchemy` to connect to SQLite.

```python
# Example: Basic Flask app with SQLite
from flask import Flask, g
import sqlite3

DATABASE = 'example.db'
app = Flask(__name__)

def get_db():
    if 'db' not in g:
        g.db = sqlite3.connect(DATABASE)
    return g.db

@app.route('/')
def index():
    db = get_db()
    db.execute('CREATE TABLE IF NOT EXISTS users (id INTEGER PRIMARY KEY, name TEXT)')
    db.execute('INSERT INTO users (name) VALUES (?)', ('Alice',))
    db.commit()
    users = db.execute('SELECT * FROM users').fetchall()
    return str(users)
```

## ‚úÖ Key Takeaways
- Flask supports SQLite
- Flask uses `sqlite3` or extensions like `Flask-SQLAlchemy`
