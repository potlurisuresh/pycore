### Without Templates (Hard Way)
```python
@app.route('/')
def home():
    return '<html><body><h1>Hello!</h1></body></html>'
```

This is messy for large HTML pages and mixes code with markup.

### With Templates (Easy Way)
Store HTML in separate files and use Jinja2 to insert dynamic data.

### What is Jinja2?
**Jinja2** is a template engine that allows :
- Insert variables: `{{ variable_name }}`,
- Use loops: `{% for item in items %}`,
- Use conditionals: `{% if condition %}`,
- Template inheritance: `{% extends "base.html" %}` in HTML files

## ğŸ“ Basic Template Rendering

### Step 1: Create HTML Template
File: `templates/home.html`
```html
<head>
    <title>{{ title }}</title>
</head>
<body>
    <h1>Welcome {{ username }}!</h1>
    <p>This is a simple template with variables.</p>
</body>
```

### Step 2: Render in Flask
```python
@app.route('/')
def home():
    return render_template('home.html', username=username, title=title)
```

The `{{ }}` placeholders are replaced with actual data.

````python
from flask import Flask, render_template

app = Flask(__name__)

# Basic template with variables
@app.route('/')
def home():
    username = 'Alice'
    title = 'Welcome to My App'
    return render_template('home.html', username=username, title=title)

# Template with user profile
@app.route('/profile/<name>')
def profile(name):
    user = {
        'name': name,
        'age': 25,
        'email': f'{name}@example.com'
    }
    return render_template('user.html', user=user)

if __name__ == '__main__':
    app.run(debug=True)
````

## ğŸ¨ Static Files (CSS, JavaScript, Images)

### Linking CSS in Templates
```html
<link rel="stylesheet" href="{{ url_for('static', filename='css/style.css') }}">
```

### Linking JavaScript
```html
<script src="{{ url_for('static', filename='js/script.js') }}"></script>
```

### Linking Images
```html
<img src="{{ url_for('static', filename='images/logo.png') }}" alt="Logo">
```

Use `url_for('static', filename='...')` to generate correct paths.

````python
from flask import Flask, render_template, url_for

app = Flask(__name__)

# Page with CSS styling
@app.route('/')
def home():
    return render_template('styled.html')

if __name__ == '__main__':
    app.run(debug=True)
````

## âŒ Common Mistakes

### Mistake 1: Wrong Folder Names
âŒ Bad: Create folders named `Templates/` or `HTML/`

### Mistake 2: Hardcoding Paths
âŒ Bad: `<link rel="stylesheet" href="/css/style.css">`
âœ… Good: `<link rel="stylesheet" href="{{ url_for('static', filename='css/style.css') }}">`

### Mistake 3: Using render_template Without Importing

### Mistake 4: Forgetting to Pass Variables
âŒ Bad: `render_template('page.html')` but template uses `{{ name }}`
âœ… Good: `render_template('page.html', name='Alice')`

## âœ… Key Takeaways
