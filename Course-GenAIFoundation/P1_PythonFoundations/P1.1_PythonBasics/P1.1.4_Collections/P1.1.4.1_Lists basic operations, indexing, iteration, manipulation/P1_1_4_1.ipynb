## üì¶ What is a List?
- A list is an ordered, **mutable** collection
- It can store multiple values of different types
- Written with square brackets `[]`
- Note: Python does not have a built-in array type. If you need arrays, use modules like NumPy.

```python
lst = [10, 20, 30, 40]
print(lst)
print(type(lst))
```

## üî¢ Indexing
- Index starts at 0
- Negative indexing starts from the end

```python
lst = [10, 20, 30, 40]
print(lst[0])   # first element
print(lst[-1])  # last element
```

## üîÅ Iteration
Use `for` loops to visit each element.

```python
lst = [10, 20, 30, 40]
for item in lst:
    print(item)
```

## üõ†Ô∏è Basic Manipulation
- Add: `append()`
- Update: assign by index
- Remove: `remove()` or `pop()`

```python
lst = [10, 20, 30]
lst.append(40)
lst[1] = 25
lst.remove(10)
print(lst)
```

## üß† Common Operations 
Use built-in functions for quick insights.

```python
lst = [5, 1, 9, 3]
print("Length:", len(lst))
print("Min:", min(lst))
print("Max:", max(lst))
print("Sorted:", sorted(lst))
```

### ‚úÖ Key Takeaways
- Python lists are ordered, mutable collections
- Indexing and iteration are core skills
- Basic manipulation includes add/update/remove
- Use built-in functions like `len()`, `min()`, `max()` for quick insights
- For true arrays, use external modules like NumPy
