## üì¶ What is a Set?
A set is an **unordered** collection of **unique** elements.

```python
numbers = {1, 2, 2, 3, 4}
print(numbers)  # duplicates removed
print(type(numbers))
```

## üîß Creating Sets
You can convert lists to sets to remove duplicates.

```python
items = ["ai", "ml", "ai", "data"]
unique_items = set(items)
print(unique_items)
```

## ‚ûï Add and Remove
Use `add()` and `remove()` (or `discard()` for safety).

```python
s = {1, 2, 3}
s.add(4)
s.remove(2)
print(s)
```

## üîÅ Set Operations
Sets support math-like operations.

```python
a = {1, 2, 3, 4}
b = {3, 4, 5}
print("Union:", a | b)
print("Intersection:", a & b)
print("Difference:", a - b)
```

## ‚úÖ Membership Test
Fast lookup with `in`.

```python
skills = {"python", "ai", "ml"}
print("python" in skills)
print("java" in skills)
```

### ‚úÖ Key Takeaways
- Sets store unique elements
- Order is not guaranteed
- Great for fast membership checks
- Supports union, intersection, difference
- Useful in data cleaning and AI workflows
