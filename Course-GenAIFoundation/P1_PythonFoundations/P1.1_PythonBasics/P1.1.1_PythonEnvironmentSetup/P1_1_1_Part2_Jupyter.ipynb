## 1. What are Cells?

A notebook is made up of **cells**. There are two main types:

### Code Cells
- Contain Python code
- Can be executed
- Show output below the cell

### Markdown Cells
- Contain formatted text (like this cell)
- Used for explanations and notes

---
## 3. How to Execute Cells

To run a code cell, use:

- **`Shift + Enter`**: Run cell and move to next cell
- **`Ctrl + Enter`**: Run cell and stay on same cell

**Try running the cell below:**

```python
# Press Shift + Enter to run this cell
print("Hello, World!")
print("You just executed your first Python code!")
```

---
## 4. Cell Execution Status

Look at the left side of code cells. You'll see:

- **`[ ]`**: Cell has not been executed yet
- **`[*]`**: Cell is currently executing (running)
- **`[1]`**: Cell has been executed (number shows execution order)

**Try running the cell below to see the execution number:**

```python
# Run this cell and watch the [*] change to a number
result = 10 + 20
print(f"The result is: {result}")
```

### For Long-Running Cells

When a cell takes time to execute, you'll see `[*]` while it's running:

```python
# This cell takes a few seconds - watch for [*]
import time

print("Starting...")
time.sleep(3)  # Wait for 3 seconds
print("Done!")
```

## 5. Your First Python Code

Let's write some simple Python code!

```python
# Hello World - The classic first program
print("Hello, World!")
```

```python
# Variables
name = "Student"
age = 25

print(f"Hello, {name}!")
print(f"You are {age} years old.")
```

### Variables Persist Between Cells

Important: Variables created in one cell are available in other cells!

```python
# Create a variable
message = "This message is stored!"
print(message)
```

```python
# Use the variable from the previous cell
print(message)  # This works because 'message' was created above
```
