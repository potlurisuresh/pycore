## Method 3 — Search: Why this method?

### 🤔 What problem did it solve?

IF–THEN rules and decision trees worked for **bounded** problems — you knew all possible inputs in advance and could write a rule for each.

But some problems have too many possibilities to enumerate:

> *"How do I get from Home to College?"* — millions of possible routes through the road network. You can't write one IF–THEN rule per route.

**Search was the answer.** Instead of pre-writing every outcome, you let the system **explore** the space of possibilities at runtime — trying paths, comparing them, and finding the best one.

So: **Search = exploring options step-by-step to reach a goal.**

### 🗺️ You've seen this in real life — Early GPS Devices

Those early GPS devices from the 2000s (like Garmin or MapMyIndia) had **no internet, no real-time traffic**. Before your trip, you entered "Home → Airport" and it calculated the route.

What it actually did: ran a search algorithm through all the roads it knew, tried different combinations, and picked the shortest path. That's Search — structured exploring at runtime, not pre-written rules for every possible journey.

> 💡 We won't code search algorithms here. Just get the core idea — *structured exploration at runtime* replaced *pre-written rules for every possible outcome*.

### The city map the code will search through

````python
# 🗺️ Search — finding the shortest route between two locations
# The GPS doesn't know the answer in advance — it explores the map step by step

# A small city map: each key connects to its neighbouring locations
city_map = {
    "Home":     ["Market", "School"],
    "Market":   ["Home", "Hospital", "Airport"],
    "School":   ["Home", "Airport"],
    "Hospital": ["Market"],
    "Airport":  ["Market", "School"]
}

def find_route(start, goal):
    """Simple BFS — explores level by level until it finds the goal."""
    queue = [[start]]          # each item is a path tried so far
    visited = set()

    while queue:
        path = queue.pop(0)
        current = path[-1]

        if current == goal:
            return path        # found it — return the full route

        if current not in visited:
            visited.add(current)
            for neighbour in city_map[current]:
                queue.append(path + [neighbour])

    return None  # no route found

# Try it
print("Home → Airport:", " → ".join(find_route("Home", "Airport")))
print("Home → Hospital:", " → ".join(find_route("Home", "Hospital")))
print("School → Market:", " → ".join(find_route("School", "Market")))
````

## Method 4 — Planning: Why this method?

### 🤔 What pushed engineers beyond single-goal search?

Search finds **one destination**. But many real tasks require a **sequence of steps** where each step depends on the previous:

> *Robot must: pick up part → move to station → place part → return to base*

You can't just search for "the next best step" — you need to reason about the whole sequence, its ordering, and its dependencies upfront.

**Planning solved this.** Where search finds a path, planning decides *what actions to take, in what order, to reach a multi-step goal*.

So:
- **Search** = *How do I get there?* (exploring paths to a single destination)
- **Planning** = *What steps do I take, in what order?* (sequencing actions toward a compound goal)

### 🏧 You've seen this in real life — the ATM Transaction Flow

Every ATM transaction follows a fixed planned sequence. Skip a step or do it out of order — the system won't work:

```
Step 1: Insert card
Step 2: Verify PIN
Step 3: Show main menu
Step 4: Customer selects transaction type
Step 5: Process the transaction
Step 6: Print receipt
Step 7: Eject card
```

A programmer planned this exact sequence in advance. The ATM doesn't "decide" the order — it just follows the plan, step by step.


### The ATM plan — fixed sequence, written in advance

<img src="plan.png" width="500"/>

Every step is pre-defined. The machine doesn't decide the order — it just follows the plan. If any step fails, the rest are skipped.


````python
# 🏧 Planning — ATM transaction sequence
# A planner defines the steps in advance; the system just executes them in order

def run_atm(card_valid, pin_correct, balance, withdrawal_amount):
    plan = [
        ("Insert card",              lambda: card_valid),
        ("Verify PIN",               lambda: pin_correct),
        ("Check balance",            lambda: balance >= withdrawal_amount),
        ("Process transaction",      lambda: True),
        ("Print receipt",            lambda: True),
        ("Eject card",               lambda: True),
    ]

    print("--- ATM Session ---")
    for step, condition in plan:
        if not condition():
            print(f"  ✋ Stopped at: '{step}'")
            return
        print(f"  ✅ {step}")
    print("  💵 Cash dispensed. Have a nice day!")

# Successful transaction
run_atm(card_valid=True, pin_correct=True, balance=5000, withdrawal_amount=2000)

print()

# Wrong PIN — plan stops early
run_atm(card_valid=True, pin_correct=False, balance=5000, withdrawal_amount=2000)

print()

# Insufficient funds
run_atm(card_valid=True, pin_correct=True, balance=500, withdrawal_amount=2000)
````

## Pros and Cons of Search

✅ **What worked**
- Handles vast spaces — no need to pre-write rules for every possible path
- Finds the best solution — algorithms like Dijkstra guarantee the shortest/fastest route
- Flexible — same search logic works for maps, puzzles, and games

❌ **Where it broke down**
- Gets too slow on complex problems — the number of possibilities explodes very quickly
- No learning — it recalculates from scratch every single time, remembers nothing
- Needs a clear goal — if the goal is vague or changes, search doesn't work

> **Bottom line:** A powerful way to navigate huge possibility spaces — but it still can't learn or adapt.


---
## Pros and Cons of Planning

✅ **What worked**
- Handles multi-step tasks — not just "get there" but "do these steps, in this order"
- Understands dependencies — knows which steps must come before others
- Enabled real automation — robots, factory lines, and scheduling systems all used this

❌ **Where it broke down**
- Still fully hand-specified — a human defines every goal, every step, every order
- No learning — if the plan fails, a human rebuilds it manually
- Brittle — one unexpected change and the whole plan needs to be rewritten

> **Bottom line:** A big leap for automation, but still fully pre-programmed. The machine follows the plan — it doesn't think for itself.

## The Big Picture — All Four Methods

Every method in this module was a genuine step forward:

- **IF–THEN Rules** → encode expert knowledge as conditions
- **Decision Trees** → organise those rules into structured branches  
- **Search** → explore vast possibility spaces to find the best path
- **Planning** → reason through ordered sequences of actions

But they all shared one fundamental weakness — **a human had to specify everything**. None of them could learn from experience or improve on their own.

🚀 **That is exactly the problem Machine Learning was designed to solve. That's where we go next.**
