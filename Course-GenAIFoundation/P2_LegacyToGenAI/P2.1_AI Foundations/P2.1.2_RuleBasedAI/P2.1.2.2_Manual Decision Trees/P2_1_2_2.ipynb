## Method 2 — Decision Trees: Why this method?

### 🤔 What pushed engineers beyond flat IF–THEN rules?

Flat IF–THEN rules work for simple checks. But real-world decisions are **layered**:

> *"Should we approve this loan?"* depends on income — **and then** on credit score — **and then** on employment history — **and then** on existing debts...

Writing 50 separate flat rules for this creates a pile where you lose track of which rules relate to each other.

**Decision trees solved this by adding structure.**

Instead of a flat pile of rules, you organise decisions into a **branching hierarchy** — each branch leads to a more specific question, only asked when the parent condition passes.

```python
# Flat rules (unmanageable at scale)
if income > 50000 and credit_score > 700 and employment_years > 2:  return "Approve"
if income > 50000 and credit_score > 700 and employment_years <= 2: return "Review"
if income > 50000 and credit_score <= 700 and region == "urban":    return "Review"
# ... and 47 more rules

# Decision tree (organised, readable — same logic, better structure)
if income > 50000:
    if credit_score > 700:            # only asked if income passes
        if employment_years > 2:      # only asked if credit passes
            return "Approve"
        else:
            return "Review"
    elif region == "urban":
        return "Review"
```

The logic is the same — the **structure** is the breakthrough.

## 🏦 Let's build it in code — Loan Approval Tree


````python
def loan_approval_decision(income, credit_score, employment_years, existing_loans, region, is_self_employed):
    # Using logical operators: and, or, not
    if (income > 80000 and credit_score > 750 and employment_years > 5 and existing_loans == 0) or (is_self_employed and income > 100000):
        return 'Approve - Premium Customer'
    elif income > 50000:
        if (credit_score > 700 and employment_years > 2) or (region in ['urban', 'metro'] and not is_self_employed):
            return 'Approve'
        elif credit_score > 650 and (region in ['urban', 'metro'] or is_self_employed) and existing_loans <= 1:
            return 'Review'
        else:
            return 'Review - Needs More Info'
    else:
        return 'Reject'

# Try different scenarios
print(loan_approval_decision(income=90000, credit_score=760, employment_years=6, existing_loans=0, region='urban', is_self_employed=False))
print(loan_approval_decision(income=120000, credit_score=680, employment_years=2, existing_loans=0, region='rural', is_self_employed=True))
print(loan_approval_decision(income=60000, credit_score=720, employment_years=3, existing_loans=1, region='metro', is_self_employed=False))
print(loan_approval_decision(income=35000, credit_score=660, employment_years=1, existing_loans=0, region='rural', is_self_employed=False))
````

---
## Pros and Cons of Decision Trees

✅ **What worked**
- More organised than flat rules — related conditions are grouped in branches, not scattered
- Easy to follow and explain — you can trace the exact path that led to a decision
- No data needed — an expert just designs the tree based on what they know

❌ **Where it broke down**
- Still hand-crafted — every branch is written by a human, nothing improves on its own
- Harder to maintain as it grows — add more conditions and the tree becomes a maze
- Can't handle anything outside the tree — unknown situations get a wrong or no answer

> **Bottom line:** A big step up from flat rules, but still fully human-programmed. The machine just follows the branches — it doesn't learn.

Trees work for structured decisions. But what if the problem has millions of possible paths — like finding a route on a map?
