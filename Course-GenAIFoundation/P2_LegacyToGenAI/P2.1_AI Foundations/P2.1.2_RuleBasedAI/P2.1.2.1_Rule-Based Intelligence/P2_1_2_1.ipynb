## Method 1 — IF-THEN Rules

### Why this method?

When humans first tried to build intelligent systems, the simplest approach was to **write down what an expert would do**.

A doctor, a banker, a technician — they all make decisions using conditions:
> *"If the patient has fever and cough, it's likely flu."*
> *"If the balance is less than the withdrawal amount, deny it."*

The idea: **encode that expert knowledge as rules**, and the machine follows them exactly.

This worked well for problems that were:
- Well-defined (clear inputs, clear outputs)
- Stable (the rules don't change often)
- Small enough to be written by hand

```
IF condition → THEN action

IF temperature > 30 → Turn ON AC
IF balance < withdrawal → Deny transaction
IF fever AND cough → Diagnose as flu
```


### 🏧 You've seen this in real life — the ATM

Every time you used an ATM, Rule-Based AI was running:

```
IF card is valid
    AND PIN is correct
    AND account balance ≥ withdrawal amount
→ Dispense cash

IF PIN is wrong 3 times → Block card
IF balance < withdrawal amount → Show "Insufficient funds"
```

That's it. A handful of flat IF-THEN rules, written by a programmer. No learning, no guessing — just logic encoded by a human expert.

<img src="atm.png" width="400"/>


### 🏧 Let's build the ATM — in code


````python
def atm(card_valid, pin_correct, balance, withdrawal_amount, pin_attempts):
    if not card_valid:
        return "❌ Card not recognised"
    if pin_attempts >= 3:
        return "🔒 Card blocked — too many wrong PIN attempts"
    if not pin_correct:
        return "❌ Wrong PIN"
    if balance < withdrawal_amount:
        return "❌ Insufficient funds"
    return "✅ Dispensing cash — please collect your card"

# Try different scenarios
print(atm(card_valid=True,  pin_correct=True,  balance=5000, withdrawal_amount=2000, pin_attempts=0))
print(atm(card_valid=True,  pin_correct=False, balance=5000, withdrawal_amount=2000, pin_attempts=2))
print(atm(card_valid=True,  pin_correct=False, balance=5000, withdrawal_amount=2000, pin_attempts=3))
print(atm(card_valid=True,  pin_correct=True,  balance=500,  withdrawal_amount=2000, pin_attempts=0))
print(atm(card_valid=False, pin_correct=False, balance=0,    withdrawal_amount=1000, pin_attempts=0))
````

💡 Every condition above is a rule a human wrote. The machine just follows them — no learning, no guessing.

## Pros and Cons of IF-THEN Rules

✅ **What worked**
- Easy to understand — you can read every rule and explain every decision
- No data needed — just write down what the expert knows
- Predictable — same input always gives the same output

❌ **Where it broke down**
- Doesn't scale — more complex problems need hundreds of rules, hard to maintain
- No learning — rules never improve on their own, a human must update them manually
- Brittle — anything not covered by a rule gets a wrong or no answer

> **Bottom line:** Great for simple, stable problems. Falls apart when the real world gets messy.
