## 1. Rule-Based AI (1950s-1980s)

### How it works:
Humans write exact IF-THEN rules. The computer follows them.

**Simple rules:**
```
IF amount > $5000 THEN require_approval()
IF login_attempts > 3 THEN lock_account()
IF balance < withdrawal_amount THEN deny()
```

**Real rules (real life):**
- ATM: IF balance < withdrawal_amount THEN deny
- Phone menu: IF user presses 1 THEN choose English

<img src="rule.png" width="600"/>

### The Problem:
- Only handles what you explicitly code
- Breaks with unexpected cases
- Hard to maintain with thousands of rules

## 2. Machine Learning (1990s-2010s)

### The Big Change:
Instead of humans writing rules → **the machine learns from data**

### How it works:
1. Feed the machine a large amount of data
2. The machine studies that data and learns from it
3. It then uses what it learned to make decisions on new data

**Example - Spam Detection:**

| Email | Result |
|-------|--------|
| Email 1: "Free offer! Click now! Urgent!" | Spam |
| Email 2: "Meeting rescheduled to 3pm" | Not Spam |
| Email 3: "Free offer inside, check attached" | Spam |

The machine studied thousands of emails like these and learned to decide on its own.

### Why it's better than rules:
- Learns from data — no manual rule writing needed
- Gets better as more data arrives
- Adapts to new spam tactics automatically

**Real use:** Gmail filters, Netflix recommendations, fraud detection, loan approval

<img src="ml.png" width="600"/>

### The Problem:
ML worked very well when the data was structured — like numbers in a table or words in an email.

But the real world is messier than that.

- It struggled when the same meaning came in different forms

What if the input is a photo? A voice recording? A sentence someone just typed?
ML didn't have a good way to make sense of these.

Something more powerful was needed.

## 3. Deep Learning (2010s-Present)

### What is it:
Deep Learning is a type of Machine Learning that uses **layers** to process data.

Instead of one step — it breaks the data into multiple layers, each understanding a little more than the previous one. 
The deeper you go, the more complex the understanding.

Think of it like reading a sentence:
- First you see letters
- Then words
- Then meaning

Deep Learning does exactly this — but with data.

Its layered approach could now make sense of understanding complex, messy, unstructured data

<img src="dl.png" width="900"/>

Deep Learning works the same way — using Artificial Neural Networks that mimic this layered structure.
Each layer passes its understanding to the next, just like neurons in the brain do.

### The True Power — Images, Audio, and Video:
Where Deep Learning truly dominates is **visual and audio data**.

**Example 1 — Object Detection in a Street Photo:**
```
Input: raw pixel data of a photo taken from a car's camera
Layer 1: Detects edges → lines, shapes, shadows
Layer 2: Detects objects → a person, a car, a traffic light
Layer 3: Understands scene → "Person crossing. Red light ahead. Slow down."
```
A photo is just millions of numbers (pixel values) — DL processes those layers and understands what's in the image.

<img src="obj.png" width="600"/>

**Real use:** This same approach powers self-driving cars (seeing the road), medical imaging (spotting tumours), and Google Lens (identifying objects with your phone camera).

---

### It Also Handles Text:
DL can also process unstructured text — though language understanding would later become its own dedicated field.

**Example 2 — Customer Phone Review:**
```
Real review: "Bought the Samsung Galaxy S24. Camera is stunning but battery 
drains so fast. App crashes every other day. Not worth the premium price."

Layer 1: Reads individual words → "stunning", "crashes", "drains", "not worth"
Layer 2: Understands phrases → "camera is stunning" = positive, "not worth" = negative
Layer 3: Decides → "Overall: Negative review"
```
No tables, no categories — just raw text. DL reads it, understands it, and classifies it.

### The Limitation:
While DL could process text, it still struggled with true **language understanding** — meaning, context, sarcasm, intent.

- "Can you pass the salt?" is a request, not a question about ability
- "That's just great" could be genuine or sarcastic depending on context

A dedicated focus on language was needed — and that's exactly what came next.

## 4. Natural Language Processing (NLP, 2014s)

### What problem are we solving?

Deep Learning is powerful — but when it reads text, it sees **words as individual pieces, not meaning**.

It can learn that certain word combinations appear together frequently.  
But it **cannot understand what a sentence actually means**.

This is the core gap — and it's called the **semantics problem**.

> **Semantics** = the actual meaning behind words, not just the words themselves.


### A realistic example

<img src="nlp.png" width="600"/>

That's exactly what NLP brings — the ability to understand **meaning, intent, and context**, not just patterns of words.


### So where does NLP fit?

Think of it this way — most of the data in the real world is text.  
Emails, reviews, messages, documents, support tickets.

Deep Learning alone can process that text. But NLP is what makes AI actually **understand** it.

Without NLP, AI would be like someone who can read every word in a sentence — but still miss the point of what was said.

NLP keeps growing by using all the tools — Rules, ML, and Deep Learning — picking the right one depending on the task.

**Real use:** Search engines, language translators, customer support bots

<img src="nlp-real.png" width="700"/>


### The Limitation — NLP still has a ceiling

NLP is excellent at understanding **meaning in a sentence or phrase** — that's its strength.

But it starts to struggle when things get bigger:

- **Long sequences** — understanding a 10-page document the same way it understands a sentence is hard. Meaning can drift across paragraphs.
- **Context across a conversation** — NLP can understand one message well, but keeping track of everything said earlier in a long chat is a challenge.
- **It understands, but cannot create** — NLP can classify, translate, and route. But if you ask it to *write something new*, it has no real answer.

> NLP taught machines to **understand** language.  
> But the next question was — can a machine also **generate** language?  
> That's exactly where Generative AI comes in.

## 5. Generative AI (2020s-Present)

NLP could understand language. But it couldn't produce anything new.

Generative AI changes that.

It doesn't just read text and classify it — it **reads, understands, and then writes back**.  
It can hold a conversation, remember context across many messages, and generate something that never existed before.

### What is the core shift?

Every technology before GenAI was built to **answer a question**:
- Is this spam or not?
- What product should I recommend?
- What does this sentence mean?

Generative AI is built to **create something new**:
- Write me a cover letter
- Summarise this document
- Explain this code to me

It's not retrieving a stored answer. It's generating a response from scratch, based on everything it has learned.

### How does it work?

GenAI is trained on **massive amounts of text** — books, articles, websites, code, conversations.

Through that training, it learns:
- How sentences are structured
- How ideas connect across paragraphs
- How tone, intent, and context shape meaning

When you give it a prompt, it doesn't search a database.  
It **predicts what should come next** — word by word — based on all of that learned knowledge.

### What it can create
- Text — emails, essays, summaries, code, explanations
- Images — DALL-E, Midjourney
- Audio and video — emerging fast

**Real use:** ChatGPT, GitHub Copilot, DALL-E, Gemini
