##  The Neuron

A **neuron** is the smallest unit of a neural network.

It receives:
- inputs (x)
- weights (w)
- bias (b)

Then applies an **activation function**.

Formula:

output  = activation( Σ(input × weight) + bias )

---
## 📥 Part 1 — Inputs (x)

Inputs are the data you feed into the neuron.

```
Predicting house price?
  x1 = size of house  (1600)
  x2 = number of beds (3)
  x3 = age of house   (4)

Detecting spam email?
  x1 = count of word 'free'  (2)
  x2 = count of word 'win'   (1)
  x3 = count of word 'click' (0)
```

Inputs are just numbers. Always.

---
## ⚖️ Part 2 — Weights (w)

Each input has a **weight** — a number that controls how important that input is.

```
High weight  →  this input matters a lot
Low weight   →  this input barely matters
Zero weight  →  this input is ignored
Negative weight → this input pushes the output down
```

Think of weights as **volume knobs** — each one controls how loud a particular input is.

```
House price example:
  w1 (size)  = 0.8   ← size matters a lot
  w2 (beds)  = 0.5   ← bedrooms matter somewhat
  w3 (age)   = -0.3  ← older house → lower price
```

**These weights are not set by you.**  
The network learns them during training — adjusting them until predictions are accurate.

<img src="images/weights_intuition.png" width="700"/>

---
## ➕ Part 3 — Bias (b)

The neuron multiplies each input by its weight, then adds them all up:

```
z  =  ((x1 × w1)  +  (x2 × w2)  +  (x3 × w3))  +  b
         ↑               ↑               ↑          ↑
      input×weight   input×weight   input×weight   bias
```

**What is the bias (b)?**

Bias is a small extra number the neuron adds to shift the result.  
Think of it as the neuron's default starting point — before any input arrives.

```
Without bias — the output is always forced to pass through zero
With bias    — the neuron can shift its output up or down freely
```

Real world analogy: bias is like a base salary.  
Even before any work is done, there is already a starting number.

---
## 🔛 Part 4 — Activation Function

The weighted sum z is then passed through an **activation function.**

**Why do we need this?**

Without an activation function, no matter how many neurons you add —  the whole thing is still just doing linear math.  
It cannot learn curves, patterns, or complexity.

The activation function **adds non-linearity** — the ability to bend and learn complex shapes.

```
No activation  →  straight line only
With activation →  curves and complex patterns
```

### The 3 Most Common Activation Functions

```
──────────────────────────────────────────────────────────────────
 Function    What it does                  Output range
──────────────────────────────────────────────────────────────────
 Sigmoid     Squashes any number to 0–1    0 to 1
             S-shaped curve
             f(z) = 1 / (1 + e^-z)

 ReLU        If z > 0 → keep it            0 to ∞
             If z ≤ 0 → output 0
             f(z) = max(0, z)

 Softmax     Turns a list of numbers       0 to 1
             into probabilities            (all values sum to 1)
──────────────────────────────────────────────────────────────────
```

> **When to use which one?**  
> That depends on *where in the network* the neuron lives — and we haven't built a network yet.  
> We'll come back to this in **P2.3.3** once the full picture is clear.

## 💻 Let's See It in Code — One Neuron From Scratch

Before using any framework, let's build a single neuron in pure Python.  
This makes the math completely visible.

````python
import math

# ── INPUTS ────────────────────────────────────────────────────────
# House: 1600 sqft, 3 beds, 4 years old
x1, x2, x3 = 1600, 3, 4

# ── WEIGHTS ───────────────────────────────────────────────────────
# These would normally be learned during training
# Here we set them manually just to see the mechanics
w1, w2, w3 = 0.8, 0.5, -0.3

# ── BIAS ──────────────────────────────────────────────────────────
b = 10

# ── WEIGHTED SUM ──────────────────────────────────────────────────
z = (x1 * w1) + (x2 * w2) + (x3 * w3) + b
print(f"Weighted Sum (z) : {z}")

# ── ACTIVATION FUNCTION — ReLU ────────────────────────────────────
# ReLU: if z is positive → keep it. If negative → make it 0.
def relu(z):
    return max(0, z)

output = relu(z)
print(f"Output after ReLU: {output}")

# ── ACTIVATION FUNCTION — Sigmoid ─────────────────────────────────
# Sigmoid: squashes any number into a value between 0 and 1
def sigmoid(z):
    return 1 / (1 + math.exp(-z))

output_sigmoid = sigmoid(z)
print(f"Output after Sigmoid: {round(output_sigmoid, 4)}  (between 0 and 1)")
````

---
## 🔍 What Just Happened?

```
Inputs         Weights         Weighted Sum          Activation     Output
──────         ───────         ────────────          ──────────     ──────
x1 = 1600  ×  w1 = 0.8      =   1280.0
x2 = 3     ×  w2 = 0.5      =      1.5
x3 = 4     ×  w3 = -0.3     =     -1.2
                  bias      =     10.0
                               ─────────
                    z       =   1290.3                →  ReLU     → 1290.3
                                                      →  Sigmoid  → ~1.0
```

One neuron. Four steps. That is the entire computation.

A neural network with millions of neurons is just this —  repeated millions of times, connected in layers.

---
## 🤔 One Neuron Is Not Enough — Why?

OK so we added an activation function. Does that fix everything?

No. One neuron — even with activation — is still just **one step of thinking.**

A person recognising a face is not doing one thing.  
They are checking edges, colours, proportions, shadows — all at the same time.  
One neuron can only do **one** of those checks.

To handle complex problems, you need **many neurons** —  
each one looking at something slightly different, and their answers combining together.
