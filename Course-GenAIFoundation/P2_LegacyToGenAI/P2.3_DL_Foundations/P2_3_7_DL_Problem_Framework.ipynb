## 🧭 The 3 Questions — Ask These First, Always

```
Question 1 — What kind of data do you have?
  Tabular (rows and columns)   →  ANN
  Images (pixels)              →  CNN
  Sequences (text, audio,
  time series, video)          →  RNN / LSTM

Question 2 — What is the task?
  Predict a number             →  Regression
  Predict a category           →  Classification

Question 3 — How much data do you have?
  Large dataset (10K+ samples) →  Train from scratch
  Small dataset                →  Use a pretrained model
```

Answer these three questions — and your algorithm choice becomes obvious.

## 📊 Question 1 — What Kind of Data?

This is the most important question. Data type drives algorithm choice.

### Tabular Data
Structured rows and columns — like a spreadsheet.

```
Size  | Beds | Age | Price
1000  |  2   |  5  | 200K
1500  |  3   |  8  | 300K
```

**Use: ANN (Artificial Neural Network)**  
Simple fully connected layers work well on structured data.

Real world: house prices, fraud detection, churn prediction, medical diagnosis

---

### Image Data
Pixels arranged in a grid — photos, scans, frames.

```
A 28×28 image = 784 pixels
Each pixel has a value (0–255)
ANN sees 784 numbers — loses all spatial information
CNN sees 28×28 grid — preserves spatial patterns
```

**Use: CNN (Convolutional Neural Network)**  
Designed to read spatial patterns — edges, shapes, textures.

Real world: face recognition, medical imaging, self-driving cars, photo classification

---

### Sequence Data
Data where **order matters** — each step depends on what came before.

```
Text:        'The cat sat on the ___'  →  next word prediction
Time series: [10, 12, 11, 13, 14, ?]  →  next value prediction
Audio:       sound waves over time    →  speech recognition
Video:       frames over time         →  action recognition
```

**Use: RNN / LSTM (Recurrent Neural Network)**  
Has memory — each step can remember what happened in previous steps.

Real world: autocomplete, translation, stock prediction, voice assistants

## 🎯 Question 2 — What Is the Task?

Once you know the data type — the task tells you what the output layer looks like.

```
──────────────────────────────────────────────────────────────────
 Task                   Output Layer         Loss Function
──────────────────────────────────────────────────────────────────
 Regression             1 neuron             MSE
 (predict a number)     no activation        Mean Squared Error

 Binary Classification  1 neuron             Binary Cross Entropy
 (yes / no, 2 classes)  sigmoid activation

 Multi Classification   N neurons            Categorical Cross Entropy
 (3+ categories)        softmax activation   (N = number of classes)
──────────────────────────────────────────────────────────────────
```

**Examples:**
```
Predict house price           →  Regression   →  1 neuron, MSE
Classify email spam/not spam  →  Binary       →  1 neuron, sigmoid
Classify image: cat/dog/bird  →  Multi (3)    →  3 neurons, softmax
Classify 10 digit types       →  Multi (10)   →  10 neurons, softmax
```

## 💾 Question 3 — How Much Data?

The amount of data determines your training strategy.

### Large Dataset — Train from Scratch
```
When:  10,000+ labeled examples (ideally much more)
How:   Build your own network, train it fully on your data
Why:   Enough signal for the network to learn from zero
```

### Small Dataset — Use a Pretrained Model
```
When:  Fewer than a few thousand labeled examples
How:   Take a model already trained on millions of examples
       Freeze the early layers (they already learned general patterns)
       Only train the last few layers on your specific data
Why:   You cannot learn from scratch with too little data
       The pretrained model already knows edges, shapes, patterns
       You are just teaching it your specific task at the end
```

This approach with pretrained models is called **Transfer Learning.**  
It is one of the most powerful ideas in modern DL.

```
Example:
  You want to classify X-rays as normal vs pneumonia
  You only have 500 labeled X-rays — not enough to train from scratch

  Solution:
  Take ResNet (pretrained on 1.2 million images)
  It already knows: edges, textures, shapes, structures
  Fine-tune the last layer on your 500 X-rays
  → Works surprisingly well
```

## 🗺️ The Full Decision Table

Put all three questions together:

```
─────────────────────────────────────────────────────────────────────────────────
 Data Type      Task                    Data Size       →  Approach
─────────────────────────────────────────────────────────────────────────────────
 Tabular        Regression              Any             →  ANN, train from scratch
 Tabular        Binary Classification   Any             →  ANN, sigmoid output
 Tabular        Multi Classification    Any             →  ANN, softmax output

 Image          Classification          Large           →  CNN from scratch
 Image          Classification          Small           →  Pretrained CNN (Transfer)
 Image          Object Detection        Any             →  Pretrained CNN

 Text           Sentiment Analysis      Any             →  RNN / LSTM
 Text           Next Word Prediction    Any             →  RNN / LSTM
 Audio          Speech to Text          Large           →  RNN / LSTM from scratch
 Audio          Speech to Text          Small           →  Pretrained model
 Time Series    Next Value Prediction   Any             →  RNN / LSTM
─────────────────────────────────────────────────────────────────────────────────
```

## 🧪 Practice — What Would You Use?

Apply the 3 questions to each problem below before reading the answer:

---

**Problem 1:**  
A hospital wants to predict whether a patient will be readmitted within 30 days.  
Data: age, diagnosis codes, number of medications, previous visits.

```
Data type?   Tabular (rows and columns)
Task?        Binary classification (readmit / not readmit)
Data size?   Hospital has 50,000 patient records — large enough

→  ANN, 1 output neuron, sigmoid, binary cross entropy loss
```

---

**Problem 2:**  
A startup wants to classify satellite images of farmland into crop types.  
They have 800 labeled images.

```
Data type?   Images
Task?        Multi-class classification (wheat/corn/rice/soy)
Data size?   800 images — small

→  Pretrained CNN (Transfer Learning)
   Use ResNet or VGG pretrained on ImageNet
   Fine-tune final layers on the 800 farm images
```

---

**Problem 3:**  
A fintech company wants to predict tomorrow's stock price based on the last 30 days.

```
Data type?   Time series (sequence — order matters)
Task?        Regression (predict a number — the price)
Data size?   Years of daily price data — large enough

→  RNN / LSTM, 1 output neuron, no activation, MSE loss
```

## ✅ Summary — What You Learned

| Concept | Key Point |
|---|---|
| Question 1 — Data type | Tabular → ANN. Image → CNN. Sequence → RNN/LSTM |
| Question 2 — Task | Regression = 1 neuron. Binary = sigmoid. Multi = softmax |
| Question 3 — Data size | Large = train from scratch. Small = pretrained model |
| Transfer Learning | Reuse a model trained on millions of examples for your small task |
| Output layer shape | Always determined by the task — not the data type |
| Loss function | Always matched to the task — MSE for regression, Cross Entropy for classification |
