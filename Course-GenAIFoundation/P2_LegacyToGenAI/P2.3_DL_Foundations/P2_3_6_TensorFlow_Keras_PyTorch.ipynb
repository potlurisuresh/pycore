## 🔷 TensorFlow — The Industrial Engine

**Built by:** Google Brain team  
**Released:** 2015  
**Used by:** Google, Airbnb, Twitter, Uber, NASA

TensorFlow was built to run Deep Learning at **massive scale** — across thousands of machines, on mobile devices, in browsers, and in production systems.

```
TensorFlow strengths:
  ✅  Production ready — deploy to servers, phones, embedded devices
  ✅  TensorFlow Lite — run models on mobile and IoT devices
  ✅  TensorFlow.js   — run models directly in the browser
  ✅  TensorFlow Serving — serve models via API in production
  ✅  Scales to massive datasets and distributed training
```

```
TensorFlow challenges:
  ⚠️  Raw TensorFlow code can be verbose and complex
  ⚠️  Steeper learning curve compared to PyTorch
  ⚠️  Debugging is harder when something goes wrong
```

> TensorFlow is the engine under the hood.  
> Most people do not use TensorFlow directly — they use **Keras**, which sits on top of it.

## 🟡 Keras — The Friendly Face of TensorFlow

**Built by:** François Chollet (at Google)  
**Released:** 2015, officially part of TensorFlow since 2019  
**Used by:** Beginners, researchers, fast prototyping

Keras is a **high-level API** that sits on top of TensorFlow.  
It wraps TensorFlow's complex operations in clean, readable code.

```
TensorFlow is the engine.
Keras is the steering wheel, dashboard, and gear system.

You drive using Keras.
TensorFlow does the heavy work underneath.
```

```
Keras strengths:
  ✅  Beginner friendly — readable, minimal code
  ✅  model.fit() handles the entire training loop
  ✅  Fast to prototype — build and test ideas quickly
  ✅  Great documentation and community
  ✅  Built into TensorFlow — no separate install needed
```

```
Keras challenges:
  ⚠️  Less control — training loop is hidden
  ⚠️  Harder to customise when you need non-standard behaviour
  ⚠️  Research teams often need more flexibility
```

**The relationship in one picture:**

```
┌───────────────────────────────────┐
│              Keras                │   ← You write code here
│   (high-level, clean, readable)   │
├───────────────────────────────────┤
│           TensorFlow              │   ← This does the actual computation
│  (low-level, GPU, backprop, ops)  │
└───────────────────────────────────┘
```

## 🔴 PyTorch — The Researcher's Choice

**Built by:** Facebook AI Research (now Meta AI)  
**Released:** 2016  
**Used by:** OpenAI, Tesla, MIT, Stanford, most cutting-edge research

PyTorch was built with a different philosophy — **flexibility first.**  
It gives you full control over every part of the training process.

```
PyTorch strengths:
  ✅  Full control — you write every step of the training loop
  ✅  Dynamic computation graph — easier to debug
  ✅  Feels like writing normal Python — intuitive
  ✅  Dominant in research — most papers use PyTorch
  ✅  Strong GPU support
```

```
PyTorch challenges:
  ⚠️  More code to write — no model.fit()
  ⚠️  More decisions to make — optimiser, loop, loss all manual
  ⚠️  Production deployment slightly more involved
```

**Why do researchers prefer PyTorch?**

```
When you are inventing a new algorithm:
  → You need to change how the forward pass works
  → You need a custom loss function
  → You need to inspect gradients at specific layers
  → You need full control of every step

PyTorch gives you that control.
Keras does not — it hides too much.
```

## 💻 The Same Network — 3 Ways

Let's write the exact same simple network in all three styles.  
Same architecture: 3 inputs → 8 hidden neurons → 1 output.

````python
# ══════════════════════════════════════════════════════════════════
# VERSION 1 — Keras (clean, minimal, automatic)
# ══════════════════════════════════════════════════════════════════
from tensorflow import keras
from tensorflow.keras import layers

model_keras = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=(3,)),
    layers.Dense(1)
])

model_keras.compile(optimizer='adam', loss='mse')

print("Keras model summary:")
model_keras.summary()
````

````python
# ══════════════════════════════════════════════════════════════════
# VERSION 2 — PyTorch (explicit, full control)
# ══════════════════════════════════════════════════════════════════
import torch
import torch.nn as nn

# Define the network architecture
model_pytorch = nn.Sequential(
    nn.Linear(3, 8),   # 3 inputs → 8 hidden neurons
    nn.ReLU(),         # Activation
    nn.Linear(8, 1)    # 8 hidden → 1 output
)

# Define loss and optimiser separately
loss_fn   = nn.MSELoss()
optimizer = torch.optim.Adam(model_pytorch.parameters(), lr=0.01)

print("PyTorch model:")
print(model_pytorch)

# Count parameters manually
total_params = sum(p.numel() for p in model_pytorch.parameters())
print(f"\nTotal parameters: {total_params}")
print("  3×8 weights + 8 biases (hidden) = 32")
print("  8×1 weights + 1 bias  (output)  = 9")
print("  Total = 41 — same as Keras above")
````

````python
# ══════════════════════════════════════════════════════════════════
# VERSION 3 — Pure Python (no framework)
# This is what we built in P2.3.3 — shown here for comparison
# ══════════════════════════════════════════════════════════════════
import math

def relu(z):    return max(0, z)
def sigmoid(z): return 1 / (1 + math.exp(-z))

# One neuron computation — the building block
def neuron(inputs, weights, bias, activation):
    z = sum(x * w for x, w in zip(inputs, weights)) + bias
    return activation(z)

print("Pure Python neuron output:", round(
    neuron([1600, 3, 4], [0.5, 0.3, -0.2], 0.1, relu), 2
))
print()
print("Compare the 3 versions:")
print("  Pure Python  →  Full transparency, no scalability")
print("  Keras        →  Clean code, automatic training loop")
print("  PyTorch      →  Full control, explicit training loop")
````
