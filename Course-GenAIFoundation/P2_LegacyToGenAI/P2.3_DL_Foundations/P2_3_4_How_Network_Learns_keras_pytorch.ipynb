## 🎯 The Learning Loop — Big Picture First

Don't worry about the details just yet. Here is the complete learning loop in plain English — just read through it once:

```
Step 1 — Forward Pass                    ← you already know this from P2.3.3!
         Feed data through the network → get a prediction

Step 2 — Measure the Error (Loss)
         Compare prediction to the real answer → how wrong are we?

Step 3 — Backpropagation
         Send the error backwards through the network
         Figure out how much each weight contributed to the error

Step 4 — Update Weights (Gradient Descent)
         Nudge every weight slightly in the direction that reduces error

Step 5 — Repeat
         Do this thousands of times until the error is small
```

Steps 2, 3, and 4 are new — and this notebook walks through each one clearly.

Here is the magic of it all:

The network starts with completely random weights.  
It makes terrible predictions at first — and that is totally expected!  
But after thousands of small corrections — it gets remarkably good.

## 📉 Step 2 — Loss — Measuring How Wrong We Are

After the Forward Pass, we compare the prediction to the real answer.  
The **Loss** is a single number that answers one simple question: **how wrong is the model right now?**

Think of it like a score on a test — except lower is better:

```
High Loss  →  predictions are far from reality  →  model needs a lot of fixing
Low Loss   →  predictions are close to reality  →  model is doing well
Loss = 0   →  perfect predictions               →  never happens in practice (and that's okay!)
```

The entire goal of training is to **minimise the loss.** Everything else is in service of this one goal.

### Two Common Loss Functions

You don't need to memorise all loss functions — just know which two to reach for:

```
────────────────────────────────────────────────────────────────
 Loss Function          When to use              What it measures
────────────────────────────────────────────────────────────────
 Mean Squared Error     Predicting a number      Average squared
 (MSE)                  (e.g. house price)       distance from truth

 Cross Entropy Loss     Predicting a category    How wrong the
                        (e.g. cat / dog / spam)  predicted probability is
────────────────────────────────────────────────────────────────
```

**MSE — Let's walk through it step by step:**
```
Actual price    = ₹300K
Predicted price = ₹340K
Error           = 340 - 300 = 40
Squared Error   = 40² = 1600

Do this for all training examples, take the average → that's your MSE
```

**Cross Entropy — Here's the intuition:**
```
Actual label    = Spam  (1)
Predicted prob  = 0.3   (only 30% confident it is spam)

The model isn't sure → that low confidence = high loss
```

## 🔙 Step 3 — Backpropagation — How Error Travels Backwards

Okay — so the Loss told us *how wrong* the network is.  
But that is not enough. We need to know: **which weights caused the problem, and by how much?**

That is exactly what **Backpropagation** figures out — and it is actually quite clever.

Once the Forward Pass produces a prediction and we compute the loss, Backprop sends that error **backwards** through the network, layer by layer:

```
Forward Pass  →  Input → Layer 1 → Layer 2 → Output → Prediction
                                                           ↓
                                                         Loss
                                                           ↓
Backprop      ←  Input ← Layer 1 ← Layer 2 ← Output ← Error
```

At each layer going backwards, it asks:  
**"How much did this weight contribute to the final error?"**

The answer to that question is called the **Gradient** — and it tells us  
exactly how much to adjust each weight, and in which direction.

> 💡 You will never have to implement backpropagation yourself.  
> Keras and PyTorch handle it automatically behind the scenes.  
> But understanding *what it does* will make you a much better practitioner.

## ⛰️ Step 4 — Gradient Descent — Using the Gradients to Fix the Weights

Backpropagation told us: *how much each weight contributed to the error, and in which direction.*

Now we use that information to **fix the weights.** This is called **Gradient Descent** — and a simple analogy makes it click instantly.

Imagine you are blindfolded on a hilly landscape and you want to reach the lowest valley.  
You can't see anything — but you *can* feel the slope under your feet.  
So you take one small step downhill. Then another. And another. Eventually you reach the bottom.

That is exactly what Gradient Descent does with weights:

```
Every point on the landscape = a different set of weight values
The height at any point      = the loss with those weights
```

```
High point  →  bad weights  →  high loss
Low point   →  good weights →  low loss
Valley      →  best weights →  minimum loss  ← this is where we want to be!
```

Gradient Descent uses the gradients from Backprop to always step **downhill** — toward lower and lower loss:

```
Start at a random point (random weights — always the starting point!)
      ↓
Backprop tells us: which direction is downhill?
      ↓
Take a small step in that direction (update the weights)
      ↓
Repeat — until we reach the valley (minimum loss)
```

## 🔢 Learning Rate — How Big is Each Step?

Each downhill step in Gradient Descent has a size — and that size is controlled by the **Learning Rate.**

Think of it like driving to a destination:

```
Learning Rate too HIGH  →  big steps  →  you overshoot the valley
                                          loss bounces around, never settles

Learning Rate too LOW   →  tiny steps →  you creep along forever
                                          training takes way too long

Learning Rate just right → steady steps → you arrive at the valley smoothly 🎯
```

Typical starting values to try: `0.01`, `0.001`, `0.0001`

> You set the learning rate when you compile your model.  
> We will see exactly where in the code — just a moment away!

## 🔄 Epochs and Batch Size

Two more terms you will see in every single training run. Let's demystify them now:

### Epoch — How many times does the model see your data?

One **Epoch** = one complete pass through the entire training dataset.

```
1 Epoch = model has seen every training example once

Training for 50 epochs = model has gone through the full dataset 50 times
```

Think of it like studying for an exam:  
Reading your notes once (1 epoch) helps. Reading them 50 times helps a lot more.  
But reading them 500 times on no sleep? That's when problems start — **overfitting** (covered in P2.3.5).

### Batch Size — How much data per update?

Instead of updating weights after every single example (which is slow and noisy),  
we feed data in small groups called **batches.**

```
1000 training examples, batch size = 32

→ 1000 / 32 = ~31 batches per epoch
→ Weights updated 31 times per epoch
→ After 50 epochs: 31 × 50 = 1550 weight updates total
```

```
Small batch size  →  noisier updates, but sometimes finds a better solution
Large batch size  →  stable updates, but needs more memory
Common default    →  32 or 64  (start here — works well most of the time)
```

## 💻 Let's See It — Training in Keras and PyTorch

Alright — theory done. Let's put it all together in actual code!  
Same simple problem we've been using: predict house price from size, bedrooms, and age.

Run both cells and watch the loss drop with each epoch. That's the network learning in real time. 🚀

````python
# ── KERAS VERSION ─────────────────────────────────────────────────
import numpy as np
from tensorflow import keras
from tensorflow.keras import layers

# ── DATA ──────────────────────────────────────────────────────────
X = np.array([
    [1000, 2, 10],
    [1200, 3,  5],
    [1500, 3,  8],
    [1800, 4,  3],
    [2000, 4,  2],
    [2300, 5,  1],
], dtype=float)

y = np.array([200, 250, 300, 360, 400, 450], dtype=float)

# ── BUILD THE NETWORK ─────────────────────────────────────────────
# Input layer:  3 features
# Hidden layer: 8 neurons, ReLU activation
# Output layer: 1 neuron (predicting a number)
model = keras.Sequential([
    layers.Dense(8, activation='relu', input_shape=(3,)),  # Hidden layer
    layers.Dense(1)                                         # Output layer
])

# ── COMPILE ───────────────────────────────────────────────────────
# Loss: MSE — we are predicting a number
# Optimizer: Adam — a smart version of Gradient Descent
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.01),
    loss='mse'
)

# ── TRAIN ─────────────────────────────────────────────────────────
# verbose=0 suppresses per-epoch output — we print it ourselves
history = model.fit(X, y, epochs=200, batch_size=4, verbose=0)

# ── LOSS CURVE ────────────────────────────────────────────────────
# Print loss every 50 epochs to see it dropping
print("=== LOSS DROPPING OVER EPOCHS ===")
for epoch in [0, 49, 99, 149, 199]:
    print(f"  Epoch {epoch+1:>3}  →  Loss: {history.history['loss'][epoch]:.2f}")

# ── PREDICT ───────────────────────────────────────────────────────
new_house = np.array([[1600, 3, 4]])
prediction = model.predict(new_house, verbose=0)
print(f"\nPredicted Price for 1600sqft, 3bed, 4yr: ₹{prediction[0][0]:.1f}K")
````

````python
# ── PYTORCH VERSION ───────────────────────────────────────────────
import torch
import torch.nn as nn

# ── DATA ──────────────────────────────────────────────────────────
X_pt = torch.tensor([
    [1000, 2, 10],
    [1200, 3,  5],
    [1500, 3,  8],
    [1800, 4,  3],
    [2000, 4,  2],
    [2300, 5,  1],
], dtype=torch.float32)

y_pt = torch.tensor([200, 250, 300, 360, 400, 450], dtype=torch.float32).view(-1, 1)

# ── BUILD THE NETWORK ─────────────────────────────────────────────
# Same architecture: 3 inputs → 8 hidden → 1 output
model_pt = nn.Sequential(
    nn.Linear(3, 8),    # Hidden layer — 3 inputs, 8 neurons
    nn.ReLU(),          # Activation
    nn.Linear(8, 1)     # Output layer — 8 inputs, 1 output
)

# ── LOSS AND OPTIMIZER ────────────────────────────────────────────
loss_fn   = nn.MSELoss()                                      # Mean Squared Error
optimizer = torch.optim.Adam(model_pt.parameters(), lr=0.01)  # Adam optimizer

# ── TRAINING LOOP ─────────────────────────────────────────────────
# In PyTorch you write the loop yourself — nothing is hidden
print("=== LOSS DROPPING OVER EPOCHS ===")
for epoch in range(200):

    # Step 1 — Forward Pass
    predictions = model_pt(X_pt)

    # Step 2 — Compute Loss
    loss = loss_fn(predictions, y_pt)

    # Step 3 — Backpropagation
    optimizer.zero_grad()   # Clear old gradients
    loss.backward()         # Compute new gradients

    # Step 4 — Update Weights
    optimizer.step()

    if epoch in [0, 49, 99, 149, 199]:
        print(f"  Epoch {epoch+1:>3}  →  Loss: {loss.item():.2f}")

# ── PREDICT ───────────────────────────────────────────────────────
new_house_pt = torch.tensor([[1600, 3, 4]], dtype=torch.float32)
with torch.no_grad():   # No gradient needed during inference
    prediction_pt = model_pt(new_house_pt)
print(f"\nPredicted Price for 1600sqft, 3bed, 4yr: ₹{prediction_pt.item():.1f}K")
````

| MSE | Used when predicting a number — measures average squared distance from truth |

| Gradient Descent | Walk downhill on the loss landscape — each step reduces the loss |
| Learning Rate | Controls step size — too high overshoots, too low is very slow |
| Backpropagation | Sends error backwards through the network to find each weight's contribution |
| Epoch | One full pass through the entire training dataset |
| Batch Size | Number of examples processed before weights are updated — default 32 or 64 |
