## 🧱 Where ML Hits a Wall — With Text Data

### Supervised: Detecting Sarcasm in Reviews

Task: Classify this review as positive or negative.

```
Review: "Oh great, another delay. Just what I needed."
```

The words say positive things — *great*, *needed*.  
But the meaning is clearly negative. It is sarcasm.

With ML (Naive Bayes / CountVectorizer) — you count words:
```
'great'   →  usually positive
'needed'  →  usually neutral
→  Model predicts: POSITIVE  ❌
```

To fix this with ML, a human would have to manually engineer features like:
```
Does the sentence start with 'Oh'?    →  add feature
Is there a comma after the first word? →  add feature
Is the tone ironic?                    →  how do you even measure this?
```

It becomes impossible. Language is too rich for hand-crafted features.

DL reads the full sentence as a sequence — word by word, in order —  
and learns that *Oh great + another delay* together signals sarcasm.

## 🧱 Where ML Hits a Wall

Let's say you want to build a system that recognises your face to unlock your phone.

With ML, the first question is: **what features do you give it?**

```
Distance between eyes?   →  measure it
Shape of nose?           →  measure it
Jawline angle?           →  measure it
Lighting condition?      →  measure it
Face turned sideways?    →  measure it again
Face with glasses?       →  measure it again
Face in the dark?        →  ...
```

You can see the problem.

**A human has to sit down and define every meaningful feature by hand.**  
This is called **Feature Engineering** — and it is the biggest bottleneck in ML.

For simple problems like house prices — manageable.  
For complex problems like faces, audio, or video — it breaks down completely.

> **Deep Learning was invented to solve exactly this.**  
> Instead of asking humans to define features — DL learns the features itself, directly from raw data.

## 🤖 What Deep Learning Does Differently

We just saw two examples where ML hits a wall — **faces** and **text**.  
The problem in both cases was the same: a **human had to sit in the middle** and define what to measure.

In ML, the pipeline always looks like this:
```
Raw Data  →  [YOU extract features manually]  →  ML Model  →  Prediction
                         ↑
              This is where ML breaks down
              (too hard for faces, too rich for language)
```

Deep Learning removes that human step entirely:
```
Raw Data  →  [System figures out what matters by itself]  →  Prediction
                                   ↑
                     No human in the middle
```

The system looks at raw pixels, raw audio waves, or raw sentences —  and works out on its own what patterns to pay attention to.

You do not tell it what to look for. It discovers that on its own.  
**That is the fundamental shift — and it is why Deep Learning unlocked everything ML couldn't.**

## 🧠 So What IS Deep Learning?

Deep Learning is a type of Machine Learning that uses **Neural Networks** —  systems loosely inspired by how the human brain processes information.

The word **"Deep"** refers to the number of layers in the network.  
More layers = deeper network = ability to learn more complex patterns.

```
Shallow Network  →  1-2 hidden layers  →  simple patterns
Deep Network     →  many hidden layers →  complex patterns
                                          (faces, speech, meaning)
```

A network with many layers is literally called a **Deep Neural Network** —  
that is where the name Deep Learning comes from.
