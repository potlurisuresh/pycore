## 🔁 Quick Recap — ML Types at a Glance

In **4.1**, we established that ML has 3 types:

| Type | How it learns | Common use |
|---|---|---|
| **Supervised** | From labeled historical data (input + known answer) | House prices, spam detection, fraud |
| **Unsupervised** | From unlabeled data — finds hidden patterns | Customer segmentation |
| **Reinforcement** | From trial and error — reward/penalty feedback | Game AI, robotics |

**Supervised Learning is by far the most common in real-world industry.**  
Why? Because almost every business already has historical data with known outcomes:
- Banks have past loan applications + whether they defaulted
- Hospitals have patient records + diagnoses
- Real estate platforms have past house sales + prices

If you have **data with answers**, Supervised Learning is your starting point.

**After training, compare Train Score vs Test Score:**

```
  Train ≈ Test  (both high)   →  ✅ Good Fit     — learned the real pattern
  Train >> Test               →  ❌ Overfitting  — memorized training data
  Train ≈ Test  (both low)    →  ❌ Underfitting — model too simple
```

| Scenario | Train R2 | Test R2 | What it means |
|---|---|---|---|
| ✅ Good Fit | 0.95 | 0.92 | Generalizes well to new houses |
| ❌ Overfitting | 0.99 | 0.55 | Memorized the training houses, fails on new ones |
| ❌ Underfitting | 0.50 | 0.48 | Missed the pattern — model too simple |
