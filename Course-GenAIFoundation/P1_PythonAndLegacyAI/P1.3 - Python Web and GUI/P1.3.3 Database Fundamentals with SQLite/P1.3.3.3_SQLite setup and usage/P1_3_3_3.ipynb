## üåê Install DB Browser for SQLite
1. Open: https://sqlitebrowser.org/
2. Download the installer for your OS
3. Run the installer and finish setup

After installation, open **DB Browser for SQLite** from your apps list.

## üß© Create or Open a Database
You can **open existing** `.db` files or **create a new one**.

**Create a new database:**
1. Click **New Database**
2. Choose a location and name (e.g., `demo.db`)
3. Click **Save**

**Open an existing database:**
1. Click **Open Database**
2. Select an existing `.db` file
3. Click **Open**

## ‚ñ∂Ô∏è Run SQL Commands in the App
1. Click the **Execute SQL** tab
2. Paste the SQL commands below
3. Click **Run** (play button)
4. View output in the bottom pane

## üëÄ Browse Data and Structure Visually
You can explore the database without writing SQL:
- **Database Structure** tab ‚Üí see tables, columns, and schema
- **Browse Data** tab ‚Üí view table rows visually

You can switch between **visual browsing** and **raw SQL** anytime.

### ‚úÖ Key Takeaways
- DB Browser for SQLite is a GUI tool for SQLite
- You can **create or open** `.db` files
- Use **Execute SQL** to run raw SQL commands

- Results appear in the bottom pane
