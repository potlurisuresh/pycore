## â“ What is a Vector Database?
A **vector database** stores data as **vectors** (lists of numbers).
These numbers represent meaning, so we can find **similar** items quickly.

## âœ… Why Do We Need Vector Databases?
Traditional databases are great for exact matches.
Vector databases are great for **similarity search**.

They help when you want to find items that are **closest in meaning**.

**Examples:**
- Find similar documents
- Recommend products
- Search images by description
- Retrieve relevant chat memory

1. Take a piece of data (e.g., a sentence)

3. Store those numbers in a **vector database**
