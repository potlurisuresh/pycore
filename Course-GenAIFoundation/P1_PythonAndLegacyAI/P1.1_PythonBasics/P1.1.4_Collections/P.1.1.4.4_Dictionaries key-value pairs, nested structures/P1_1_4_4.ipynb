## üì¶ What is a Dictionary?
A dictionary stores data as **key-value** pairs.
Keys are unique and used to access values quickly.

```python
student = {"name": "Aisha", "age": 22, "course": "AI"}
print(student)
print(type(student))
```

## üîç Accessing Values
Use the key to access its value.

```python
student = {"name": "Aisha", "age": 22, "course": "AI"}
print(student["name"])
print(student.get("course"))
```

## üõ†Ô∏è Add / Update / Delete
Dictionaries are mutable and can be changed.

```python
student = {"name": "Aisha", "age": 22}
student["course"] = "AI"   # add
student["age"] = 23         # update
del student["name"]          # delete
print(student)
```

## üîÅ Iteration
Iterate through keys, values, and items.

```python
student = {"name": "Aisha", "age": 22, "course": "AI"}
for key in student:
    print(key)

for value in student.values():
    print(value)

for key, value in student.items():
    print(key, value)
```

## üß© Nested Dictionaries (Intermediate)
Dictionaries can store other dictionaries.

```python
students = {
    "s1": {"name": "Aisha", "score": 91},
    "s2": {"name": "Ravi", "score": 88}
}
print(students["s1"]["name"])
print(students["s2"]["score"])
```

## üß† Common Patterns
- Count frequencies
- Group values
- Fast lookup using keys

```python
text = "ai ai python data ai"
freq = {}
for word in text.split():
    freq[word] = freq.get(word, 0) + 1
print(freq)
```
