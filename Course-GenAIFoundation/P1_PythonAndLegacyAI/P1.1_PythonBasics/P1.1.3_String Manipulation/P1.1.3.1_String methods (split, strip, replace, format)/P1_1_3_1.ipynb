## ‚úÇÔ∏è split()
- Break a string into a list using a separator
- Common in CSV, logs, and user input

```python
text = "python,ai,ml"
parts = text.split(",")
print(parts)
```

## üßº strip()
- Removes whitespace from both ends of a string
- Useful for cleaning input data

```python
raw = "   Hello World   "
clean = raw.strip()
print(clean)
```

## üîÅ replace()
- Replaces a substring with another substring
- Safe for simple text edits

```python
msg = "I love Java"
updated = msg.replace("Java", "Python")
print(updated)
```

## üß© format()
- Inserts values into a template string
- Good for readable output

```python
name = "Aisha"
score = 92
print("{} scored {}%".format(name, score))
```

## ‚úÖ Quick Practice (Intermediate)
Combine methods for cleaner data

```python
raw = "  python | ai | ml  "
items = raw.strip().split("|")
items = [x.strip() for x in items]
print(items)
```

## ü§ñ AI Use Case (Brief)
These methods are core to text preprocessing:
- `strip()` cleans noisy input
- `split()` tokenizes text
- `replace()` normalizes terms
- `format()` builds clean prompts and logs

### ‚úÖ Key Takeaways
- `split()` turns text into lists
- `strip()` removes outer whitespace
- `replace()` edits strings safely
- `format()` (and f-strings) create clean output
- Essential for text cleaning, logging, and AI pipelines
