## ðŸ§© What is String Interpolation?
String interpolation inserts variables into strings.
Pythonâ€™s best option: **f-strings**.

```python
name = "Maigha"
score = 92
print(f"{name} scored {score}%")
```

## âœ… Expressions Inside f-strings
You can use math or logic directly inside `{}`.

```python
price = 250
qty = 3
print(f"Total: {price * qty}")
```

## ðŸ”¢ Number Formatting
Control decimal places, commas, and percentages.

```python
value = 12345.6789
print(f"Rounded: {value:.2f}")
print(f"With commas: {value:,.2f}")
print(f"Percentage: {0.8734:.1%}")
```

## ðŸ“… Formatting Dates (Intermediate)

```python
from datetime import datetime
now = datetime(2026, 2, 3)
print(f"Today: {now:%Y-%m-%d}")
print(f"Readable: {now:%b %d, %Y}")
```

## ðŸ§  Multi-line f-strings (Intermediate)

```python
user = "Ravi"
level = "Intermediate"
report = f"""
User: {user}
Level: {level}
Status: Active
"""
print(report)
```

## âš¡ f-strings vs format() (Quick Comparison)
- f-strings are shorter and faster
- `format()` is still useful in older codebases

```python
name = "Lina"
score = 88
print(f"{name} scored {score}%")
print("{} scored {}%".format(name, score))
```
