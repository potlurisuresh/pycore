## üßµ What is a String?
- A string is a sequence of characters
- Used to represent text data
- Strings are immutable (cannot be changed in-place)

```python
name = "Python"
message = 'Welcome to AI'
print(name)
print(type(name))
```

## üî° Creating Strings
- Single quotes `' '`
- Double quotes `" "`
- Triple quotes for multi-line text

```python
single = 'Hello'
double = "World"
multi = '''This is
a multi-line
string'''
print(single)
print(double)
print(multi)
```

## üîó String Concatenation and Repetition

```python
first = "Python"
second = "AI"
print(first + " " + second)
print(first * 3)
```

## üìè String Length and Indexing
- Index starts from 0
- Negative indexing is supported

```python
text = "Artificial"
print(len(text))
print(text[0])
print(text[-1])
```

## ‚úÇÔ∏è String Slicing
`string[start:end]`

```python
word = "Intelligence"
print(word[0:4])
print(word[:5])
print(word[5:])
```

## üõ†Ô∏è Common String Methods (Very Important)
- lower(), upper()
- strip()
- replace()
- split()
- join()

```python
sentence = "  Learning Python is Fun  "
print(sentence.lower())
print(sentence.strip())
print(sentence.replace("Fun", "Powerful"))
print(sentence.split())
```

## üß© String Formatting
Modern and clean formatting using f-strings

```python
name = "Maigha"
score = 95
print(f"{name} scored {score}% in Python")
```

## ü§ñ Real-World Use Case: Text Preprocessing (AI)
Strings are heavily used in NLP and data cleaning.

```python
review = "  This Course is AMAZING!!!  "
clean_review = review.lower().strip()
print(clean_review)
```

### ‚úÖ Key Takeaways
- Strings represent text data in Python and are immutable, meaning any modification creates a new string.
- Python provides powerful built-in string methods (lower, strip, split, replace, join) to clean and transform text efficiently.
- Indexing and slicing allow precise access to parts of a string, which is essential for parsing and data processing.
- Strings are core to real-world applications and AI, especially in text preprocessing, APIs, and Natural Language Processing (NLP).
