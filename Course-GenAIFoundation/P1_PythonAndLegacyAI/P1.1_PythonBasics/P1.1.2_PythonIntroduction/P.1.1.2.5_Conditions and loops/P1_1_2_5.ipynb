## âœ… Conditional Statements (if / elif / else)
Conditions let your program make decisions based on boolean expressions.

```python
# Basic if/else
score = 72
if score >= 60:
    print("Pass")
else:
    print("Fail")
```

### ðŸ” Multiple Conditions with elif
Use `elif` for multiple branches.

```python
# Grading example
score = 85
if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
else:
    print("Grade: D or below")
```

### âœ… Nested Conditions (Intermediate)
You can place conditions inside other conditions for layered decisions.

```python
# Nested if example
age = 20
has_id = True

if age >= 18:
    if has_id:
        print("Access granted")
    else:
        print("ID required")
else:
    print("Access denied")
```

## ðŸ” For Loops
Use `for` to iterate over sequences (strings, lists, ranges, etc.).

```python
# Loop through a list
languages = ["Python", "Java", "C++"]
for lang in languages:
    print(lang)
```

### ðŸ”¢ Using range() in for loops

```python
# Print numbers 1 to 5
for i in range(1, 6):
    print(i)
```

### ðŸ§  For Loop with Condition

```python
# Filter even numbers
numbers = [1, 2, 3, 4, 5, 6]
for n in numbers:
    if n % 2 == 0:
        print(f"Even: {n}")
```

## â™»ï¸ While Loops
Use `while` when you donâ€™t know how many times youâ€™ll repeat.

```python
# Countdown using while
count = 5
while count > 0:
    print(count)
    count -= 1
print("Done")
```

## ðŸ›‘ Loop Control: break and continue
- `break` exits the loop
- `continue` skips the current iteration

```python
# break and continue example
for i in range(1, 6):
    if i == 3:
        continue  # skip 3
    if i == 5:
        break     # stop at 5
    print(i)
```
