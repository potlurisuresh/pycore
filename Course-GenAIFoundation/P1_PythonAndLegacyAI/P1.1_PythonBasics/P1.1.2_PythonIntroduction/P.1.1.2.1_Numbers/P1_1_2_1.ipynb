### ðŸ”¢ Integer (int)
- Represents whole numbers
- Can be positive or negative
- No decimal point


```python
a = 10
b = -5
print(a, b)
print(type(a))
```

### ðŸ”¢ Float (float)
- Represents decimal numbers
- Used when precision is required
- Very common in data science & AI

```python
x = 10.5
y = 3.14
print(x, y)
print(type(x))
```

### ðŸ”¢ Complex Numbers

- `b` â†’ imaginary part
- `j` represents âˆš(-1)

```python
c1 = 3 + 4j
c2 = complex(2, 5)

print(c1)
print(c2)
print(type(c1))
```
